Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = df[df['Язык.программирования'] == 'Python']
python_data
```

Визначте розмір таблиці python_data за допомогою методу shape

```python
python_data.shape
```

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
python_data = python_data.groupby('Должность')

python_data
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
aggregated_data = python_data.agg({'Зарплата.в.месяц': ['min', 'max']})
aggregated_data
```

Cтворіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(data):
    return (data[0] + data[1]) / 2
```

```python
aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
aggregated_data
```

```python
plt.figure(figsize=(7, 7))
plt.pie(
    aggregated_data['avg'],
    labels=aggregated_data['avg'].keys(),
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)
plt.title("Середня заробітня плата за посадами", loc='left')
plt.show()
```
