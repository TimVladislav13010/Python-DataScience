```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
```

```python
tmp = pd.read_html(url, match="Коефіцієнт народжуваності в регіонах України")
```

```python
# print(tmp[0])
table = tmp[0]
```

```python
table.head()
```

# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
table.shape
```

# Замініть у таблиці значення "—" на значення NaN

```python
table =  table.replace("—", np.nan)
```

```python
table
```

# Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
table.dtypes
```

# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
for column in table.columns:
#     print(table[column])
    if column not in "Регіон":
        table[column] = table[column].astype(np.float64)
#         table[column]
        
table.dtypes
```

```python
table
```

# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
print((table.isnull().sum())/len(table))
```

# Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
table = table.drop(table.index[-1])
table
```

# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
table = table.fillna({i: table[i].mean() for i in table if i != 'Регіон'})
table
```

```python
plt.barh(table['Регіон'], table['2019'])
plt.xlabel("Коефіцієнт", fontsize="small", color="midnightblue")
plt.ylabel("Регіон", fontsize="small", color="midnightblue")
plt.title("Коефіцієнт народжуваності у регіонах України у 2014р (народжень на 1000 осіб, ‰)", fontsize=15)
plt.show()
```

# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
average_bd = table["2019"].mean()
average_bd
```

```python
average_table_bd = table[table['2019'] > average_bd][['Регіон', '2019']]
average_table_bd
```

```python
plt.pie(
    average_table_bd['2019'],
    labels=average_table_bd['Регіон'],
    shadow=True,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
)

plt.show()
```

# У якому регіоні була найвища народжуваність у 2014 році?

```python
average_bd_max_1_2014 = table["2014"].max()
average_bd_max_1_2014
```

```python
average_bd_max_1_2014 = table[table['2014'] == table["2014"].max()][['Регіон', '2014']]
average_bd_max_1_2014
```

# Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
y = table['2019']
x = table['Регіон']
plt.bar(x, y, label="2019")
plt.xlabel('Регіони')
plt.xticks(rotation=90)
plt.ylabel('Коефіцієнт народжуваності')
plt.legend()
plt.show()
```
